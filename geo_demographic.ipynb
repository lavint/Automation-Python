```python
from uszipcode import SearchEngine
search = SearchEngine()  # add simple_zipcode=False to get more info
```

```python
zipcode = search.by_zipcode('10001')
nyc = zipcode.to_dict()
resultset = {key: value for key, value in nyc.items() if key not in {'polygon'}}
resultset
```
